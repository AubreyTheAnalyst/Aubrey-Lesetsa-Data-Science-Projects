## Plotting a bar graph to visualize data

```python
plt.bar(df['Flavor'],df['Flavor Rating'],color='skyblue')
plt.xlabel('Cookie Flavor')
plt.ylabel('Flavor rating in %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
