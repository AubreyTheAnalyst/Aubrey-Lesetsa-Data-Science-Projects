## Gold Price increase per ounce over the last 10 years

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

## Loading Data

```python
df =pd.read_csv('gold_price_data.csv')
last_10_yrs = df.head(10)
```

## Selecting data from only the last 10 years

```python
last_10_yrs
```

## Checking missing data and datatype

```python
last_10_yrs.isnull().sum()
last_10_yrs.describe
last_10_yrs.dtypes
```

## Sorting date

```python
date_sorted= last_10_yrs.sort_values(by='Date')
date_sorted
```

## Visualizing the data

```python
date_sorted.plot(x='Date',y='Annual Average Price',kind='line',color= 'orange')
plt.title('Price of gold over the last 10 years')
plt.xlabel('Date in years')
plt.ylabel('Price in Dollars($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Price of gold in 2016: $1251.92

```python
_2016_gold_price= date_sorted[0:1]
_2016_gold_price
```

## Price of gold in 2025: $3124.05

```python
_2025_gold_price= date_sorted[9:10]
_2025_gold_price
```

## Gold price difference between 2016 and 2025

```python
price_increase_since_2016 = date_sorted['Annual Average Price'].max() - date_sorted['Annual Average Price'].min()
print(f'The increase in the price of gold from 2016 to 2025 is: {price_increase_since_2016}')
```

**The increase in gold price from 2016 to 2025 is $1872.13.**
